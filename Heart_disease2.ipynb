```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load dataset
datapath = "D:\heart_disease_classifacation\heart-disease-prediction_ca9d91fb-0255-41f2-9e92-d5cd33edca38.csv"
df = pd.read_csv(datapath)
print(df.head())

# Step 2: Outlier handling [Skipped]

# Step 3: Check and handle missing values
print("\nMissing values:\n", df.isnull().sum())
df = df.dropna()

# Step 4: Separate features and target
X = df.drop("TenYearCHD", axis=1)
y = df["TenYearCHD"]

# Step 5: Skip Standardization

# Step 6: Feature selection (select top 10 features)
selector = SelectKBest(score_func=f_classif, k=10)
X_selected = selector.fit_transform(X, y)
selected_features = X.columns[selector.get_support()]
print("\nSelected Features:\n", selected_features)

# Step 7: Split data
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Step 8: Train logistic regression model
if len(np.unique(y_train)) < 2:
    print("Error: y_train contains only one class. Model training skipped.")
    model = None
else:
    model = LogisticRegression(max_iter=1000, solver='liblinear')
    model.fit(X_train, y_train)

# Step 9: Predict and evaluate
if model is not None:
    y_pred = model.predict(X_test)

    print("\nAccuracy:", accuracy_score(y_test, y_pred))
    print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("\nClassification Report:\n", classification_report(y_test, y_pred, zero_division=0))

    # Display feature coefficients
    print("\nModel coefficients (Feature: Coefficient):")
    for feature, coef in zip(selected_features, model.coef_[0]):
        print(f"{feature}: {coef:.4f}")
else:
    print("Model was not trained. Skipping evaluation and coefficient display.")

# Step 10: Visualize confusion matrix
if model is not None:
    plt.figure(figsize=(6, 4))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='coolwarm')
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title("Confusion Matrix")
    plt.tight_layout()
    plt.show()
else:
    print("Confusion matrix cannot be displayed as the model was not trained.")
```
